### Расчет RR

Создадим функции для расчета Retention Rate.

```python
# функция для создания пользовательских профилей
def get_profiles(users, payments):
    # находим параметры первых посещений
    profiles = users
    # для когортного анализа определяем дату первого посещения
    # и первый день месяца, в который это посещение произошло
    profiles['dt'] = profiles['reg_time'].dt.date
    profiles['month'] = profiles['reg_time'].astype('datetime64[M]')
    # добавляем признак платящих пользователей
    profiles['payer'] = profiles['user_id'].isin(payments['user_id'].unique())
    # считаем количество уникальных пользователей с одинаковой датой привлечения
    new_users = (
        profiles.groupby('dt')
        .agg({'user_id': 'nunique'})
        .rename(columns={'user_id': 'unique_users'})
        .reset_index()
    )
    return profiles
```

```python
# функция для расчёта удержания
def get_retention(
    profiles,
    sessions,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):
    # добавляем столбец payer в передаваемый dimensions список
    dimensions = ['payer'] + dimensions
    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')
    # собираем «сырые» данные для расчёта удержания
    result_raw = result_raw.merge(
        sessions[['user_id', 'open_time']], on='user_id', how='left'
    )
    result_raw['lifetime'] = (
        result_raw['open_time'] - result_raw['reg_time']
    ).dt.days

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        result = result.div(result['cohort_size'], axis=0)
        result = result[['cohort_size'] + list(range(horizon_days))]
        result['cohort_size'] = cohort_sizes
        return result

    # получаем таблицу удержания
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)
    # возвращаем таблицу и сырые данные
    return result_raw, result_grouped
```

```python
# функция для визуализации удержания
def plot_retention(retention, horizon):
    # задаём размер сетки для графиков
    plt.figure(figsize=(10, 5))
    # исключаем размеры когорт и удержание первого дня
    retention = retention.drop(columns=['cohort_size', 0])
    # если в индексах таблицы удержания только payer,
    # добавляем второй признак — cohort
    if retention.index.nlevels == 1:
        retention['cohort'] = 'All users'
        retention = retention.reset_index().set_index(['cohort', 'payer'])
    # в таблице графиков — два столбца и одна строка, всего две ячейки
    # в первой строим кривые удержания платящих пользователей
    ax1 = plt.subplot(1, 2, 1)
    retention.query('payer == True').droplevel('payer').T.plot(
        grid=True, ax=ax1
    )
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание платящих пользователей')
    # во второй ячейке строим кривые удержания неплатящих
    # вертикальная ось — от графика из первой ячейки
    ax2 = plt.subplot(1, 2, 2, sharey=ax1)
    retention.query('payer == False').droplevel('payer').T.plot(
        grid=True, ax=ax2
    )
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание неплатящих пользователей')
    plt.tight_layout()
    plt.show()
```

Зададим момент и горизонт анализа и построим кривые удержания платящих и неплатящих пользователей.

```python
# задаем горизонт анализа и момент анализа
observation_date = datetime(2023, 5, 22).date()  # момент анализа
horizon_days = 10  # горизонт анализа
# получаем профили пользователей
profiles = get_profiles(users, payments)
# считаем удержание
retention_raw, retention = get_retention(
    profiles, sessions, observation_date, horizon_days, dimensions=[]
)
# строим графики
plot_retention(retention, horizon_days)
```

**Выводы**

Удержание платящих пользователей выше на любой лайфтайм. На обеих графиках происходит плавное снижение к 10-му лайфтайму: примерно 50% платящих пользователей и менее 10% неплатящих пользователей возвращаются на 10-й день. Для платящих пользователей можно отметить необычный всплеск возвратов на 3-ий день (кривая удержания в этот лайфтайм выше, чем в предыдущий).

### Расчет времени сессии

```python
# добавим столбец с длительностью в датафрейм с сессиями
sessions['session_duration_sec'] = (sessions['close_time'] - sessions['open_time']).dt.seconds
# строим гистограмму
sessions['session_duration_sec'].hist(bins=50)
plt.title('Распределение длительности сессий')
plt.xlabel('Длительность сессии, сек')
plt.show()
```

На графике видно большое количество значений около нуля - это сесии пользователей, завершенные из-за технической ошибки. Также есть несколько выбросов с выоскими значениями (до 22 часов и более) - это аномально длинные сессии. Чтобы расчитать среднюю продолжительность сесии верно, такие выбросы нужно удалить.

```python
# определили 99-е перцентили
q_low = sessions['session_duration_sec'].quantile(0.01)
q_hi  = sessions['session_duration_sec'].quantile(0.99)
# очистили данные от выбросов по длительности сессии и отфильтровали датафрем
sessions_filtered = sessions[(sessions['session_duration_sec'] < q_hi) & (sessions['session_duration_sec'] > q_low)]
#
print(f'Удалили строк: {((sessions.shape[0] - sessions_filtered.shape[0]) / sessions.shape[0]):.2%}')
```

```python
print('Медианная длительность сессии:', sessions_filtered['session_duration_sec'].median())
print('Средняя длительность сессии:', round(sessions_filtered['session_duration_sec'].mean(),2))
```

Медианная длительность сессии около семи минут. Средняя - около 11.

**Выводы**

Медианная длительность сессии около семи минут. Средняя - около 11. Однако, чтобы оценить «здоровье» сессий, медианную длительность сессии нужно сравнить с ожидаемой.

Если интерес пользователей к приложению будет падать, то будет падать удержание и средннее время сесии. Это, а также то, что пользователи не хотят платить за продукт, приводит к снижению метрик монетизации (ARPU, ARPPU). Поэтому эти метрики стоит отслеживать в первую очередь.

Метрики монетизации показывают, сколько выручки приносят пользователи, но не показывают, сколько денег тратится на их привлечение/удержание, и, соответственно, не показывают, сколько прибыли мы получаем с пользователей. Поэтому следует также отслеживать стоимость привлечения пользователя и окупаемость рекламы в виде метрик CAC и ROI. Для их расчета необходимы данные о рекламных кампаниях.
