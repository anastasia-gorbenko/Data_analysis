**About Dataset**

The dataset is built from the initial dataset consisted of 600000 transactional data collected in 6 years (period 2014-2019), indicating date and time of sale, pharmaceutical drug brand name and sold quantity, exported from Point-of-Sale system in the individual pharmacy. Selected group of drugs from the dataset (57 drugs) is classified to the following Anatomical Therapeutic Chemical (ATC) Classification System categories:

- M01AB - Anti-inflammatory and antirheumatic products, non-steroids, Acetic acid derivatives and related substances
- M01AE - Anti-inflammatory and antirheumatic products, non-steroids, Propionic acid derivatives
- N02BA - Other analgesics and antipyretics, Salicylic acid and derivatives
- N02BE/B - Other analgesics and antipyretics, Pyrazolones and Anilides
- N05B - Psycholeptics drugs, Anxiolytic drugs
- N05C - Psycholeptics drugs, Hypnotics and sedatives drugs
- R03 - Drugs for obstructive airway diseases
- R06 - Antihistamines for systemic use

Sales data are resampled to the hourly, daily, weekly and monthly periods. Data is already pre-processed, where processing included outlier detection and treatment and missing data imputation.
****

**Описание данных** 

Данные взяты из одной розничной аптеки в Нише, Сербия. Датасет построен на основе первоначального набора данных, состоящего из 600000 транзакций, собранных за 6 лет (2014-2019 гг.), с указанием:
- даты и времени продажи;
- торговой марки фармацевтического препарата;
- количества проданных, экспортированных из системы точек продаж в отдельную аптеку. 

Группа препаратов из набора данных (57 препаратов) относится к следующим категориям анатомо-терапевтической классификации (АТХ):

- M01AB - Противовоспалительные и противоревматические продукты, нестероиды, производные уксусной кислоты и связанные с ними вещества
- M01AE - Противовоспалительные и противоревматические продукты, нестероиды, производные пропионовой кислоты
- N02BA - Другие анальгетики и жаропонижаютические препараты, салициловая кислота и производные
- N02BE/B - Другие анальгетики и жаропонижаютические препараты, пиразолоны и анилиды
- N05B - Психолептические препараты, анксиолитические препараты
- N05C - Психолептические препараты, гипнотические и cедативные препараты
- R03 - Препараты от обструктивных заболеваний дыхательных путей
- R06 - Антигистаминные препараты для системного применения

По данным сделана повторная выборка в почасовой, дневной, недельный и ежемесячный периоды. **Данные уже предварительно обработаны. Обработка включала обнаружение и обработку выбросов и восстановление или удаление пропусков**.

### Предобработка данных

Данные уже предварительно обработаны. Обработка включала обнаружение и обработку выбросов и восстановление или удаление пропусков. Осталось только поменять тип данных для даты и переименовать столбцы для облегчения дальнейшей работы.

````python
# выполнили переименование столбцов
data.columns = data.columns.str.lower()
data = data.rename(columns = {'datum':'date', 'weekday name':'weekday_name'})
````

````python
# привели столбец с датой продажи к типу datetime
data['date'] = pd.to_datetime(data['date'], format='%m/%d/%Y')
display(data.head())
````

Продажи по каждой группе препаратов можно сагрегировать еще глубже, чтобы найти дополнительные закономерности. Например, большинство препаратов групп M01AB, M01AE, N02BA, N02BE/B могут использоваться в качестве обезболивающих и жаропонижающих.

N05B и N05C используются как успокоительные и снотворные средства.

R03 (препараты для лечения обструктивных заболеваний дыхательных путей) и R06 (антигистаминные препараты для системного применения) используются при аллергии и бронхиальной астме.

````python
# сгруппировали столбцы с продажами препаратов, которые могут использоваться 
# при одном и том же состоянии, и добавили столбцы с суммой
data['analgetic'] = data['m01ab'] + data['m01ae'] + data['n02ba'] + data['n02be']
data['sedative'] = data['n05b'] + data['n05c']
# добавили столбец с номером дня недели продажи
data['weekday_number'] = data['date'].dt.weekday
# добавили столбец с номером дня в месяце
data['day'] = data['date'].dt.day
# добавили столбец с общими продажами
data['total_sales'] = data['analgetic'] + data['sedative'] + data['r03'] + data['r06']
data.head()
````

````python
# посмотрели на распрделение количества проданных в день препаратов
data_for_boxplot = data[['m01ab', 'm01ae', 'n02ba', 'n02be', 'n05b', 'n05c', 'r03', 'r06', 'analgetic', \
                   'sedative']]
data_for_boxplot.boxplot(figsize=(10, 5))
plt.ylim(-5, 90)
plt.show()
````

**Вывод**

У каждой группы препаратов бывают дни с всплесками продаж.

Из представленных групп количество продаж выше у групп N02BE (анальгетики и жаропонижаютические препараты, пиразолоны и анилиды), N05B (успокоительные) и R03 (от обструктивных заболеваний дыхательных путей)

### Исследовательский анализ данных

#### Исследование общего количества продаж

````python
# применили стиль для графиков
plt.style.use('seaborn-darkgrid')
# создали функцию для рисования гистограмм
def histogram(df, period, values, xlabel, title):
    df.pivot_table(index=period, values=values, aggfunc='sum').plot(kind='bar', figsize=(7,3), legend=False)
    plt.xlabel(xlabel, fontsize=14, loc ='right')
    plt.ylabel('Количество проданных лекарств, уп', fontsize=12)
    plt.title(title, fontsize=16)
    plt.show()
````

````python
# посмотрели информацию по количеству продаж в 2019 году
histogram(data.query('year == 2019'), 'month', 'total_sales', 'Месяц', 'Количество проданных лекарств за 2019 год по месяцам')
````

За 2019 год данные не полные: не хватает данных, начиная с октября по декабрь. Для корректного сравнения продажи за 2019 лучше исключить.

````python
# сделали срез по годам
good_data = data.query('year != 2019')
# посмотрели информацию по количеству продаж по годам
histogram(good_data, 'year', 'total_sales', 'Год', 'Количество проданных лекарств по годам')
````

````python
# посмотрели информацию по количеству продаж по месяцам
histogram(good_data, 'month', 'total_sales', 'Месяц', 'Количество проданных лекарств по месяцам')
````

````python
# посмотрели информацию по количеству продаж по дням
histogram(good_data, 'day', 'total_sales', 'День месяца', 'Количество проданных лекарств по дням месяца')
````

````python
# посмотрели информацию по количеству продаж по дням недели
histogram(good_data, 'weekday_number', 'total_sales', 'День недели, где 0 - пн, 1 - вт и т.д.',\
          'Количество проданных лекарств по дням недели')
````

**Вывод**

- Общие продажи лекарств данных групп росли с 2014 по 2016 год, в 2017 продажи упали, в 2018 начали снова расти.
- Общие продажи лекарств данных групп заметно выше в осенне-зимний период: продажи растут с августа, достигая пика в октябре, с февраля начинаетсся спад, который достигает минимума в летние месяцы.

#### Проверка гипотезы о средних продажах антигистаминных в мае 2017 и 2018 года

```Основная гипотеза: средние количества проданных препаратов группы r06 в мае 2017 и в мае 2018 года равны.```

```Альтернативная гипотеза: средние количества проданных препаратов группы r06 в мае 2017 и в мае 2018 года равны.```

Для проверки гипотезы используем t-тест.

````python
# results = вызов метода для проверки гипотезы
s1 = good_data[(good_data['year'] == 2017) & (good_data['month'] == 5) ]['r06']
s2 = good_data[(good_data['year'] == 2018) & (good_data['month'] == 5) ]['r06']
results = st.ttest_ind(s1, s2)
# alpha = задайте значение уровня значимости
alpha = 0.05
# вывод значения p-value на экран 
print(results.pvalue)
# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

**Вывод**

На данном уровне значимости есть основая отвергнуть гипотезу о том, что средние количества проданных препаратов группы r06 в мае 2017 и в мае 2018 года равны. 

С практической точки это значит, что вероятность того, что в мае 2018 года продажи антигистаминных препаратов оказались выше, чем в 2017 году, всего 2,4%. Скорее всего, продажи увеличились не случайно.
